```python
import json
import random

# File paths
vulnerable_file = "vulnerables.json"
non_vulnerable_file = "non-vulnerables.json"
output_file = "dataset.json"

# Load vulnerable data
with open(vulnerable_file, "r", encoding="utf-8") as f:
    vulnerable_data = json.load(f)

# Add label = 1 to vulnerable samples
for item in vulnerable_data:
    item["label"] = 1

# Load non-vulnerable data
with open(non_vulnerable_file, "r", encoding="utf-8") as f:
    non_vulnerable_data = json.load(f)

# Add label = 0 to non-vulnerable samples
for item in non_vulnerable_data:
    item["label"] = 0

# Combine datasets
combined_data = vulnerable_data + non_vulnerable_data

# Shuffle randomly
random.seed(42)
random.shuffle(combined_data)

# Save combined dataset
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(combined_data, f, indent=2)

print(f"Combined dataset saved to {output_file}")
print(f"Total samples: {len(combined_data)}")
print(f"Vulnerable: {len(vulnerable_data)}")
print(f"Non-Vulnerable: {len(non_vulnerable_data)}")
```
